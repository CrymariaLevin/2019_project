```python
import math

#math.sqrt(x)

def distance_calc(l1,l2):
    l_minus = []
    for i in range(len(l1)):
        li = l1[i]-l2[i]
        l_minus.append(li)
    sum_l = sum(x*x for x in l_minus)
    distance = math.sqrt(sum_l)
    return distance

l1 = [5,6,7,8]
l2 = [1,2,3,4]
distance_calc(l1,l2)
```
