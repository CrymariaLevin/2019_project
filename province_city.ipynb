```python
# 

import pymysql
import pandas
from bs4 import BeautifulSoup
import requests
import numpy as np

header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 6.1; WOW64; rv:53.0) Gecko/20100101 Firefox/53.0"
    }

def get_mobile_location(num): #
    url = 'http://www.ip138.com:8080/search.asp?mobile={}&action=mobile'.format(str(num))
    web_origin = requests.get(url, headers=header).content
    soup = BeautifulSoup(web_origin, 'html.parser')
    num_form = soup.find_all('td', attrs={'class':'tdc2'})
    num_location = num_form[1].get_text()
    province, city = num_location.split('\xa0')
    return province, city

def get_com_cover(name):
    connection = pymysql.connect(host = '47.92.25.70',user = 'root',password = 'Wfn031641',db = 'cxd_data',charset = 'utf8')
    with connection.cursor() as cursor: 
        sql = "select `operate` from `dic_company_detail` where `name` = '{}'".format(name)
        cursor.execute(sql)
        cover = cursor.fetchone()[0]
    connection.close()
    return cover

dir_object = 'E://.xls'
df = pandas.read_excel(dir_object, sheet_name = '')
#df.head()
for index in df.index:
    num = df.loc[index, '']
    try:
        prov, city = get_mobile_location(num)
    except:
        continue
    df.loc[index, ''] = prov
    df.loc[index, ''] = city
    
df2 = pandas.read_excel(dir_object, sheet_name = '')
for index in df2.index:
    name = df2.loc[index, '']
    try:
        cover = get_com_cover(name)
    except Exception as e:
        print(str(e))
        continue
    df2.loc[index, ''] = cover
    
writer=pandas.ExcelWriter('E://.xls')
df2.to_excel(writer, index=False, sheet_name='')
df.to_excel(writer, index=False, sheet_name='')
writer.save()
print('Done!')
```
