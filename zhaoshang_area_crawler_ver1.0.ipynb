```python
def get_map_message(name):
    key = 'e84aae1ede212a3e043dfd7e8a21cf7e'
    url = 'https://restapi.amap.com/v3/geocode/geo?parameters'
    params = {
            "key" : key, 
            "address" : name,
            "output" : 'json' 
        }
    data = requests.get(url,params = params).json()
    count = data['count']
    if count == '0':
        return None
    else:
        data_detail = data['geocodes'][0]
        city = data_detail['city']
        address = data_detail['formatted_address']
        coordinate = data_detail['location']
        data_row = [city, address, coordinate]
        return data_row

for index in df_sql.index:
    connection = connection = pymysql.connect(host = '47.92.25.70',user = 'root',password = 'Wfn031641',db = 'zhaoshang',charset = 'utf8')
    name = df_sql.loc[index, 'name']
    data = get_map_message(name)
    print(data)
    if data is None or len(data[0]) == 0:
        print(name, '无地址信息')
        continue
    else:
        data_row = data + [name,]
        #print(data_row)
        with connection.cursor() as cursor:
            sql_update = "update `dic_park` SET `city`=%s,`address`=%s,`coordinate`=%s WHERE `name`=%s"
            cursor.execute(sql_update,data_row)
        connection.commit() 
        connection.close()
        print('更新完成测试库')
```
